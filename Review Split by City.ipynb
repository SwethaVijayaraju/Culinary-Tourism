```python
def get_business_ids(places):
    cities = dict()
    for place in places:
        cities[place[0]] = set(df_business[(df_business.city == place[0]) & (df_business.state == place[1])].business_id.tolist())
    return cities
```

```python
cities = get_business_ids(places)
```

```python
def write_to_file(city, line):
    with open('/Users/swetha/Downloads/yelp_dataset/reviews_by_city/'+city+'.json', 'a+') as fp:
         fp.write(line)
```
