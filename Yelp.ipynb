## POS Tagging

```python
%%capture
# spaCy stuff
# !pip install spacy
!python -m spacy download en_core_web_sm
!python -m spacy download en

import spacy 

nlp = spacy.load("en_core_web_sm")
```

```python
import re 
import string 
import nltk
import numpy as np

from tqdm import tqdm  
from spacy.matcher import Matcher 
from spacy.tokens import Span 
from spacy import displacy 
```

```python
nltk.download("stopwords")
from nltk.corpus import stopwords
stopWords = set(stopwords.words('english'))
stopWords
```

```python
def cleanWord(string):
    pattern = re.compile(r'[a-z]+')
    word = string.lower()
    #if word in stopWords:
       # return ""
    if not word.isalpha():
        word = pattern.findall(word)
        #print(word)
    if type(word) == str:
        return word
    if type(word) == list and len(word) != 0:
        word = word[0]
        #if word in stopWords:
        #    return ""
        return word
    else:
        return ""

df_review_small['review_words'] = df_review_small.review.apply(lambda r : r.split())
df_review_small['clean_review_words'] = df_review_small.review_words.apply(lambda r : [cleanWord(word) for word in r])
df_review_small['clean_review'] = df_review_small.clean_review_words.apply(lambda r : " ".join(r))
```

```python
df_review_small
```

```python
def extract_adjectives(doc):
    adjectives = set()
    for tok in doc: 
        if tok.pos_ == "ADJ":
            adjectives.add(tok.text)
    return adjectives

def extract_nouns(doc):
    nouns = set()
    for tok in doc: 
        if tok.pos_ == "NOUN":
            nouns.add(tok.text)
    return nouns
```

```python
df_review_small["adjectives"] = df_review_small.clean_review.apply(lambda r : extract_adjectives(nlp(r)))
```

```python
df_review_small["nouns"] = df_review_small.clean_review.apply(lambda r : extract_nouns(nlp(r)))
```

```python
df_review_small
```

```python
def get_adj_noun_pairs(row):    
    text = df_review_small.review[row]
    doc = nlp(text)

    adjectives = df_review_small.adjectives[row]

    noun_chunks = []
    for chunk in doc.noun_chunks:
        noun_chunks.append(chunk.text.lower())

    adj_noun_chunks = dict()
    for adj in adjectives:
        for chunk in noun_chunks:
            if adj in chunk:
                if not adj in adj_noun_chunks:
                    adj_noun_chunks[adj] = [chunk]
                else:
                    adj_noun_chunks[adj].append(chunk)

    return adj_noun_chunks
```

```python
def get_noun_pairs(row):    
    text = df_review_small.review[row]
    doc = nlp(text)

    nouns = df_review_small.nouns[row]

    noun_chunks = []
    for chunk in doc.noun_chunks:
        noun_chunks.append(chunk.text.lower())
    
    noun_pairs = dict()
    for noun in nouns:
        for chunk in noun_chunks:
            if noun in chunk:
                if not noun in noun_pairs:
                    noun_pairs[noun] = [chunk]
                else:
                    noun_pairs[noun].append(chunk)

    return noun_pairs
```

```python
df_review_small["adj_noun_pairs"] = [get_adj_noun_pairs(i)for i in range(df_review_small.shape[0])]
```

```python
df_review_small["noun_pairs"] = [get_noun_pairs(i)for i in range(df_review_small.shape[0])]
```
