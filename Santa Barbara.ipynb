```python
def cleanWord(string):
    pattern = re.compile(r'[a-z]+')
    word = string.lower()
    #if word in stopWords:
       # return ""
    if not word.isalpha():
        word = pattern.findall(word)
        #print(word)
    if type(word) == str:
        return word
    if type(word) == list and len(word) != 0:
        word = word[0]
        #if word in stopWords:
        #    return ""
        return word
    else:
        return ""

df_merge['review_words'] = df_merge.review.apply(lambda r : r.split())
df_merge['clean_review_words'] = df_merge.review_words.apply(lambda r : [cleanWord(word) for word in r])
df_merge['clean_review'] = df_merge.clean_review_words.apply(lambda r : " ".join(r))
```

## POS tagging

```python
def extract_adjectives(doc):
    adjectives = set()
    for tok in doc: 
        if tok.pos_ == "ADJ":
            adjectives.add(tok.text)
    return adjectives

def extract_nouns(doc):
    nouns = set()
    for tok in doc: 
        if tok.pos_ == "NOUN":
            nouns.add(tok.text)
    return nouns
```

```python
df_merge_rest["adjectives"] = df_merge_rest.clean_review.apply(lambda r : extract_adjectives(nlp(r)))
```

```python
df_merge_rest["nouns"] = df_merge_rest.clean_review.apply(lambda r : extract_nouns(nlp(r)))
```

```python
df_merge_rest
```

```python
df_merge_rest.reset_index(inplace = True)
df_merge_rest.drop(columns = ["index", "level_0"], axis = 1, inplace = True)
df_merge_rest.head(3)
```

```python
def get_adj_noun_pairs(row):    
    text = df_merge_rest.review[row]
    doc = nlp(text)

    adjectives = df_merge_rest.adjectives[row]

    noun_chunks = []
    for chunk in doc.noun_chunks:
        noun_chunks.append(chunk.text.lower())

    adj_noun_chunks = dict()
    for adj in adjectives:
        for chunk in noun_chunks:
            if adj in chunk:
                if not adj in adj_noun_chunks:
                    adj_noun_chunks[adj] = [chunk]
                else:
                    adj_noun_chunks[adj].append(chunk)

    return adj_noun_chunks
```

```python
def get_noun_pairs(row):    
    text = df_merge_rest.review[row]
    doc = nlp(text)

    nouns = df_merge_rest.nouns[row]

    noun_chunks = []
    for chunk in doc.noun_chunks:
        noun_chunks.append(chunk.text.lower())
    
    noun_pairs = dict()
    for noun in nouns:
        for chunk in noun_chunks:
            if noun in chunk:
                if not noun in noun_pairs:
                    noun_pairs[noun] = [chunk]
                else:
                    noun_pairs[noun].append(chunk)

    return noun_pairs
```

```python
df_merge_rest["adj_noun_pairs"] = [get_adj_noun_pairs(i)for i in range(df_merge_rest.shape[0])]
```

```python
df_merge_rest["noun_pairs"] = [get_noun_pairs(i)for i in range(df_merge_rest.shape[0])]
```
